```python
class lr:
    def __init__(self):
        self.m = None
        self.b = None

    def train(self, X_train, y_train):
        num = 0
        den = 0

        # Get the mean scalar values
        x_mean = X_train.mean().values[0]
        y_mean = y_train.mean().values[0]

        for i in range(X_train.shape[0]):
            # Use .iloc[i, 0] to safely grab the value at row i, column 0
            x_i = X_train.iloc[i, 0]
            y_i = y_train.iloc[i, 0]

            # Correct OLS formulas
            num = num + (x_i - x_mean) * (y_i - y_mean)
            den = den + (x_i - x_mean) ** 2

        self.m = num / den
        self.b = y_mean - (self.m * x_mean)

    def predict(self, X_test):
        return self.m * X_test + self.b
```

```python
lr1 = lr()
lr1.train(X_train,y_train)
```
