```python
import numpy as np
import pandas as pd
```

```python
x = np.array([[1,1],
             [2,3],
              [3,2]])
print(x.shape)
```

```python
mean = np.mean(x,axis=0)
mean
```

```python
x_centered = x - mean
x_centered
```

```python
cov_matrix = np.cov(x_centered,rowvar=False)
cov_matrix
```

```python
eigen_values,eigen_vector = np.linalg.eigh(cov_matrix)
eigen_values,eigen_vector
```

```python
best_direction = np.argmax(eigen_values)
```
